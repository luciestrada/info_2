```python
import os
import pydicom
import matplotlib.pyplot as plt
```

```python


def cargar(carpeta):
    archivos = []
    for archivo in sorted(os.listdir(carpeta)): # Ordenar 
        if archivo.endswith('.dcm'):
            path = os.path.join(carpeta, archivo)# 1. Ruta de la carpeta donde estan los archivos, 2. Nombre del archivo que queremos acceder
            archivo_dcm = pydicom.dcmread(path) # Leer la informacion DICOM del path y guarda en la variable
            archivos.append((archivo_dcm, archivo)) #1. contenido del DICOM, 2. nombre del archivo DICOM
    return archivos

def visualizar(imagenes):
    for i, (img, nombre) in enumerate(imagenes, start=1):
        plt.imshow(img.pixel_array, cmap=plt.cm.gray) # Escala de grises
        plt.axis('off')  
        plt.title(f"Imagen {i}: {nombre}")  
        plt.show()
        input("----Presiona Enter para mostrar la siguiente imagen----")
        plt.close()

ruta = r'C:\Users\lucia\Downloads\archivosDCM'
imagenes = cargar(ruta)
visualizar(imagenes)
```
